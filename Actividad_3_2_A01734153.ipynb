```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **2020**

## **Preparación**

```python
#Leer archivo y crear 4 df distintos por año
df_2020= pd.read_excel('G&C_LIMPIA.xlsx',sheet_name="2020")
```

```python
#Verificamos información del DataFrame
df_2020.info()
```

```python
#Corroboramos valores nulos
valores_nulos=df_2020.isnull().sum()
valores_nulos
```

```python
#Creamos los dos dataframes separado valores string
columnas_object = df_2020.select_dtypes(exclude=['float', 'int'])
df_2020_str = pd.DataFrame(columnas_object)
columnas_numericas = df_2020.select_dtypes(include=['float', 'int'])
df_2020_num = pd.DataFrame(columnas_numericas)
```

## **Desviación estándar para eliminar outliers en DataFrame**


```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df_2020_num
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers del Dataframe
outliers= df_2020_num[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df_2020_num2= df_2020_num[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_2020_num2
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=df_2020_num2.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=df_2020_num2.copy()
data_clean=data_clean.fillna(round(df_2020_num2.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Finalmente concatenamos el dataframe numerico con el de string y guardamos en Excel 
df_2020_concat = pd.concat([data_clean, df_2020_str], axis=1)
df_2020_concat.to_excel('G&C_2020_Desv.xlsx', index=False)
```

## **Cuantiles para sustituir outliers en Dataframe**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df_2020_num

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= df_2020_num[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
df_2020_iqr= df_2020_num[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df_2020_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=df_2020_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
df_2020_iqr_final=df_2020_iqr.copy()
df_2020_iqr_final=df_2020_iqr_final.fillna(round(df_2020_iqr.mean(),1))
df_2020_iqr_final
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=df_2020_iqr_final.isnull().sum()
valores_nulos
```

```python
#Finalmente concatenamos el dataframe numerico con el de string y guardamos en Excel 
df_2020_concat_2 = pd.concat([df_2020_iqr_final, df_2020_str], axis=1)
df_2020_concat_2.to_excel('G&C_2020_IQR.xlsx', index=False)
```
