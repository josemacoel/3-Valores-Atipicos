```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
df= pd.read_csv('ventastotales_sin_nulos.csv')
df=df.drop("Unnamed: 0", axis=1)
df
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

## **ventas_precios_corrientes**

### **MÉTODO DE IQR (RANGO INTER QUARTILICO)**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers
outliers_iqr= df[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
df_clean_iqr_ventas_precios_corrientes= df[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df_clean_iqr_ventas_precios_corrientes
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(df_clean_iqr_ventas_precios_corrientes["ventas_precios_corrientes"]) 
plt.title("Outliers de variable")
plt.show() #dibujamos el diagrama
```

### **MÉTODO DE DESVIACIÓN ESTÁNDAR**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Outliers
outliers= df[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
df_clean= df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_clean
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(df_clean["ventas_precios_corrientes"]) 
plt.title("Outliers ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```

### **Análisis**

Para la columna de ventas_precios_corrientes se encontró un valor que es outlier con el metodo de Rango Inter Cuartílico, con el método de desviación estándar no, ya que permite mayor amplitud en los datos.
